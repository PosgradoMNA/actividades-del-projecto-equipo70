### Fundamentos de bases de datos y para ciencia de datos.

Una base de datos es una colección organizada de datos, lógicamente relacionados. Para el apartado de Ciencia y analítica de datos se hablará de bases de datos relacionales SQL. La información se organiza por tablas, es un conjunto de columnas y filas (o campos o registros); a la intersección entre columnas y filas se le conoce como celda. ### Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.

Data Warehouse es una base de datos grande que contiene información de diferentes fuentes y las reúne; de esta manera, se procesa la información para usarla con otras tecnologías. De ahí existen procesos para cargar la información a estos sistemas, como ETL o ELT (sus siglas significan extraer, transformar y cargar o extraer, cargar y transformar).  Mediante frameworks, lenguajes de programación, se pueden llevar procesos de extracción, procesamiento y carga de información.
